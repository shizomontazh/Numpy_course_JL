### Coding Task 1 — Read a row

Given:
```python
import numpy as np

data = np.array([
    [5, 7, 9],
    [2, 4, 6],
    [1, 3, 8]
])
```

Write code that stores the **second row** in a variable called `row_result`.

> **AUTO-CHECK IDEA:** compare `row_result` with `np.array([2, 4, 6])`


````python
import numpy as np

data = np.array([
    [5, 7, 9],
    [2, 4, 6],
    [1, 3, 8]
])

# Write your code here
row_result = data[1]
row_result
````

### Coding Task 2 — Read a column

Using the same array `data`, store the **last column** in a variable called `col_result`.

> **AUTO-CHECK IDEA:** compare with `np.array([9, 6, 8])`


````python
import numpy as np

data = np.array([
    [5, 7, 9],
    [2, 4, 6],
    [1, 3, 8]
])

# Write your code here
col_result = data[:, -1]
col_result
````

### Coding Task 3 — Row totals

Given:
```python
scores = np.array([
    [10, 20, 30],
    [15, 15, 15],
    [8, 12, 20]
])
```

Write code that stores the total of each row in a variable called `totals`.

> **AUTO-CHECK IDEA:** compare with `np.array([60, 45, 40])`


````python
import numpy as np

scores = np.array([
    [10, 20, 30],
    [15, 15, 15],
    [8, 12, 20]
])

# Write your code here
totals = scores.sum(axis=1)
totals
````

### Coding Task 4 — Update a block

Given:
```python
grid = np.zeros((5, 5), dtype=int)
```

Write code that changes the middle 3×3 block to `4`.  
Store the final array in the same variable `grid`.

> **AUTO-CHECK IDEA:** compare with an expected 5×5 array


````python
import numpy as np

grid = np.zeros((5, 5), dtype=int)

# Write your code here
grid[1:4, 1:4] = 4
grid
````

### Coding Task 5 — Count matching values

Given:
```python
canvas = np.array([
    [0, 255, 255],
    [100, 255, 0],
    [255, 100, 100]
])
```

Write code that stores the number of cells equal to `255` in a variable called `bright_count`.

> **AUTO-CHECK IDEA:** compare `bright_count` with `4`


````python
import numpy as np

canvas = np.array([
    [0, 255, 255],
    [100, 255, 0],
    [255, 100, 100]
])

# Write your code here
bright_count = np.sum(canvas == 255)
bright_count
````

### Coding Task 6 — Build from rules

Create a 4×4 integer array called `board` with these rules:
- start with zeros,
- fill the border with `1`,
- set the center 2×2 block to `7`.

> **AUTO-CHECK IDEA:** compare with a fixed expected array


````python
import numpy as np

# Write your code here
board = np.zeros((4, 4), dtype=int)
board[0, :] = 1
board[-1, :] = 1
board[:, 0] = 1
board[:, -1] = 1
board[1:3, 1:3] = 7
board
````

### Final Challenge — Arena map

Write NumPy code that creates a 7×7 integer array called `arena`.

Rules:
1. start with zeros,
2. fill the whole border with `2`,
3. place a horizontal wall of `5`s across row 3 from columns 1 to 5,
4. place a treasure `9` at position `(1, 5)`,
5. place a player `8` at position `(5, 1)`.

Then compute and store:
- `arena_sum` — the sum of all values,
- `wall_count` — how many cells are equal to `5`,
- `border_top` — the first row of the array.

> **AUTO-CHECK IDEA**  
> Check:
> - exact array shape,
> - exact array contents,
> - `arena_sum`,
> - `wall_count`,
> - `border_top`.


````python
import numpy as np

# Write your code here
arena = np.zeros((7, 7), dtype=int)

arena[0, :] = 2
arena[-1, :] = 2
arena[:, 0] = 2
arena[:, -1] = 2

arena[3, 1:6] = 5
arena[1, 5] = 9
arena[5, 1] = 8

arena_sum = arena.sum()
wall_count = np.sum(arena == 5)
border_top = arena[0]

print(arena)
print("arena_sum =", arena_sum)
print("wall_count =", wall_count)
print("border_top =", border_top)
````
