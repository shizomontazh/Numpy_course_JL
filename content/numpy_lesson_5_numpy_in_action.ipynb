## Scenario 1. Game scores over several days

Imagine that a player gets points in four rounds each day.

We can store one week of scores in a 2D array:
- each **row** = one day,
- each **column** = one round.


```python
import numpy as np

scores = np.array([
    [12, 18, 15, 20],
    [10, 22, 17, 19],
    [14, 16, 18, 21],
    [20, 19, 16, 15],
    [11, 25, 20, 24]
])

scores
```

### Reading the data

Before doing any calculations, make sure you understand the structure.


```python
print("Shape:", scores.shape)
print("Number of days:", scores.shape[0])
print("Rounds per day:", scores.shape[1])
```

### Quick task 1

Write code that prints:
1. all scores from Day 1,
2. the score from Day 3, Round 4,
3. all scores from Round 2.


```python
# Try it here

print("Day 1:", scores[0])
print("Day 3, Round 4:", scores[2, 3])
print("Round 2:", scores[:, 1])
```

### Summaries

This is where arrays become useful very quickly.

You can ask questions like:
- What was the total score on each day?
- Which round usually gave the most points?
- What is the highest score in the whole table?


```python
daily_totals = scores.sum(axis=1)
round_totals = scores.sum(axis=0)

print("Total points per day:", daily_totals)
print("Total points per round:", round_totals)
print("Highest score in the table:", scores.max())
print("Average score:", scores.mean())
```

### Why `axis` matters

For a 2D array:
- `axis=1` means "work across each row",
- `axis=0` means "work down each column".

That is why:
- `scores.sum(axis=1)` gives one total **per day**,
- `scores.sum(axis=0)` gives one total **per round**.


### Quick task 2

Use NumPy to find:
1. the best day total,
2. the average score for each round,
3. the difference between the best and worst day totals.


```python
best_day_total = daily_totals.max()
round_averages = scores.mean(axis=0)
difference = daily_totals.max() - daily_totals.min()

print("Best day total:", best_day_total)
print("Average score in each round:", round_averages)
print("Difference between best and worst day totals:", difference)
```

## Scenario 2. Updating a grid

Arrays are also useful when you want to **change** data in a structured way.

Imagine a 5×5 game map:
- `0` = empty cell,
- `1` = wall,
- `9` = treasure.


```python
game_map = np.zeros((5, 5), dtype=int)

game_map[1, 1:4] = 1
game_map[3, 0:3] = 1
game_map[4, 4] = 9

game_map
```

### Reading a grid

This array is small, but the same idea appears in:
- games,
- pixel art,
- maps,
- seating plans,
- puzzle boards.

Let's inspect a few parts of it.


```python
print("Top row:", game_map[0])
print("Center cell:", game_map[2, 2])
print("Bottom-right cell:", game_map[4, 4])
print("Rows 1 to 3, columns 1 to 3:")
print(game_map[1:4, 1:4])
```

### Quick task 3

Modify the map so that:
1. the whole first column becomes walls,
2. the cell at row 0, column 4 becomes treasure,
3. the middle row becomes all zeros again.

Try it yourself first, then compare with the example answer.


```python
map_copy = game_map.copy()

map_copy[:, 0] = 1
map_copy[0, 4] = 9
map_copy[2, :] = 0

map_copy
```

### A useful habit: `.copy()`

When you want to test changes without destroying the original array, use `.copy()`.

Without `.copy()`, it is easy to overwrite data and lose the version you wanted to keep.


## Scenario 3. A mini pixel canvas

A grayscale image can be represented by numbers:
- small numbers = darker pixels,
- larger numbers = brighter pixels.

We do not need real image files to understand the idea.
A simple 2D array is enough.


```python
canvas = np.array([
    [0,   0,   0,   0,   0,   0],
    [0, 255, 255, 255, 255,   0],
    [0, 255,  50,  50, 255,   0],
    [0, 255, 255, 255, 255,   0],
    [0,   0,   0,   0,   0,   0]
])

canvas
```

### Looking for patterns

Even without plotting anything, you can still reason about this as a pixel grid.

Questions you can answer:
- What is the shape of the image?
- How many bright pixels are there?
- Can you darken a region?
- Can you draw a border?


```python
print("Shape:", canvas.shape)
print("Brightest value:", canvas.max())
print("Darkest value:", canvas.min())

bright_pixels = np.sum(canvas == 255)
print("Number of bright pixels:", bright_pixels)
```

### Quick task 4

Create a new version of the canvas where:
1. all `255` values become `200`,
2. the inside 2×2 block becomes `100`,
3. the outer border stays unchanged.


```python
edited = canvas.copy()

edited[edited == 255] = 200
edited[1:3, 2:4] = 100

edited
```

### Combining ideas

In that one example, you used:
- comparison,
- boolean selection,
- slicing,
- copying,
- updating a rectangular region.

That is the point of this lesson: **real tasks usually mix several small skills together**.


## Mini investigation. Weather table

Suppose this table stores temperatures for three cities across four days.


```python
temps = np.array([
    [21, 23, 22, 24],
    [18, 19, 21, 20],
    [25, 27, 26, 28]
])

temps
```

### Investigation tasks

Use the array to answer these questions:
1. Which city had the highest single temperature?
2. What is the average temperature for each city?
3. What is the average temperature for each day?
4. Which city had the biggest temperature increase from Day 1 to Day 4?


```python
city_averages = temps.mean(axis=1)
day_averages = temps.mean(axis=0)
increase = temps[:, 3] - temps[:, 0]

print("Highest single temperature:", temps.max())
print("Average temperature for each city:", city_averages)
print("Average temperature for each day:", day_averages)
print("Increase from Day 1 to Day 4:", increase)
print("City with biggest increase (0-based index):", np.argmax(increase))
```

## Final challenge. Build and analyze a training board

Create an 6×6 integer array that represents a training board.

Rules:
1. start with zeros,
2. fill the whole border with `1`,
3. place a 2×2 square of `5`s in the center,
4. put a `9` in the top-right corner,
5. print the final board,
6. print:
   - the sum of all values,
   - the largest value,
   - the middle four cells.

Try it on your own before opening the example solution.


```python
board = np.zeros((6, 6), dtype=int)

board[0, :] = 1
board[-1, :] = 1
board[:, 0] = 1
board[:, -1] = 1

board[2:4, 2:4] = 5
board[0, -1] = 9

print(board)
print("Sum:", board.sum())
print("Largest value:", board.max())
print("Middle four cells:")
print(board[2:4, 2:4])
```
