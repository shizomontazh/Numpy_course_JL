### Q1
What does `np.arange(2, 10, 2)` return?

A. `[2, 4, 6, 8, 10]`  
B. `[2, 4, 6, 8]`  
C. `[2, 3, 4, 5, 6, 7, 8, 9]`  
D. `[10, 8, 6, 4, 2]`

**Answer:** B

---

### Q2
Which function is best when you want exactly 5 evenly spaced values from 0 to 1?

A. `np.arange(0, 1, 5)`  
B. `np.linspace(0, 1, 5)`  
C. `np.zeros(5)`  
D. `np.random.randint(0, 1, 5)`

**Answer:** B

---

### Q3
Can an array with 12 elements be reshaped into shape `(3, 5)`?

A. Yes  
B. No

**Answer:** B

---

### Q4
What does `np.arange(12).reshape(3, 4).shape` return?

A. `(12,)`  
B. `(4, 3)`  
C. `(3, 4)`  
D. `12`

**Answer:** C

---

### Q5
Which statement about `np.linspace(0, 10, 6)` is correct?

A. It creates values with step size 6  
B. It creates 6 evenly spaced values from 0 to 10  
C. It excludes the value 10  
D. It only works with integers

**Answer:** B

---

### Q6
What does `np.arange(9).reshape(3, -1)` do?

A. It creates an error  
B. It reshapes into 3 rows and lets NumPy find the number of columns  
C. It deletes one value  
D. It creates random values

**Answer:** B

---

### Q7
Which line creates a `2 x 3` array filled with ones?

A. `np.ones(2, 3)`  
B. `np.ones((2, 3))`  
C. `np.array(ones((2, 3)))`  
D. `np.fill((2, 3), 1)`

**Answer:** B

---

### Q8
What is true about `np.random.randint(1, 7, size=(2, 2))`?

A. It always gives the same numbers  
B. It creates floats  
C. It creates integers from 1 up to but not including 7  
D. It includes 7 every time

**Answer:** C
