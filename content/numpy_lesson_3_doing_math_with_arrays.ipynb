## 2. Array math is elementwise

When two arrays have the same shape, NumPy can combine them **element by element**.

That means:
- first with first
- second with second
- third with third
- and so on

Let's see it.

````python
a = np.array([2, 4, 6, 8])
b = np.array([1, 3, 5, 7])

print("a:", a)
print("b:", b)
print("a + b:", a + b)
print("a - b:", a - b)
print("a * b:", a * b)
````

### Quick check

Look at the result of `a + b`.  
Each position was calculated separately:

- `2 + 1`
- `4 + 3`
- `6 + 5`
- `8 + 7`

That is what **elementwise addition** means.

## 3. Lists and arrays do not behave the same way

This is one of the most important ideas in the lesson.

With Python lists:
- `+` joins lists together
- `*` repeats the list

With NumPy arrays:
- `+` adds values elementwise
- `*` multiplies values elementwise

````python
python_list = [1, 2, 3]
numpy_array = np.array([1, 2, 3])

print("Python list + Python list:", python_list + python_list)
print("Python list * 2:", python_list * 2)

print("NumPy array + NumPy array:", numpy_array + numpy_array)
print("NumPy array * 2:", numpy_array * 2)
````

### Why this matters

If you want to do math with many numbers, arrays are much more convenient.

A list is a general Python container.  
An array is a structure built for numeric work.

## 4. Scalar operations

A **scalar** is just one number, like `2`, `10`, or `0.5`.

When you add, subtract, multiply, or divide an array by a scalar, NumPy applies that operation to every element.

````python
scores = np.array([12, 15, 9, 18])

print("scores:", scores)
print("scores + 1:", scores + 1)
print("scores - 2:", scores - 2)
print("scores * 3:", scores * 3)
print("scores / 3:", scores / 3)
````

## 5. Combining operations

You can also build expressions.

NumPy will apply the operations across the whole array.

````python
energy = np.array([4, 7, 10, 13])

print("energy:", energy)
print("energy * 2 + 1:", energy * 2 + 1)
print("(energy + 3) / 2:", (energy + 3) / 2)
````

### Important note

In expressions like `energy * 2 + 1`, NumPy still works elementwise.

So if `energy` starts with `4`, then the first result is:

`4 * 2 + 1 = 9`

## 6. Comparisons with arrays

You can compare arrays with numbers or with other arrays.

The result is usually an array of `True` and `False` values.

````python
temps = np.array([17, 21, 19, 25, 16])

print("temps:", temps)
print("temps > 20:", temps > 20)
print("temps <= 19:", temps <= 19)
````

### Reading comparison results

Each value in the result tells you whether the condition is true for that position.

For `temps > 20`:
- `17 > 20` → `False`
- `21 > 20` → `True`
- `19 > 20` → `False`
- `25 > 20` → `True`
- `16 > 20` → `False`

## 7. Simple aggregations

Sometimes you do not want a whole new array.  
Sometimes you want one summary number.

NumPy includes many aggregation functions. In this lesson we focus on four:

- `sum()`
- `mean()`
- `min()`
- `max()`

````python
daily_steps = np.array([3500, 6200, 5400, 7100, 4800])

print("daily_steps:", daily_steps)
print("sum:", daily_steps.sum())
print("mean:", daily_steps.mean())
print("min:", daily_steps.min())
print("max:", daily_steps.max())
````

### What these mean

- `sum()` adds everything
- `mean()` gives the average
- `min()` finds the smallest value
- `max()` finds the largest value

These are some of the most useful tools for quick analysis.

## 8. Aggregations on 2D arrays

Aggregations also work with 2D arrays.

If you use them in the simplest way, NumPy looks at **all values in the array**.

````python
grid = np.array([
    [4, 8, 2],
    [7, 1, 9]
])

print("grid:")
print(grid)
print("sum:", grid.sum())
print("mean:", grid.mean())
print("min:", grid.min())
print("max:", grid.max())
````

At this stage, that is enough.  
Later, students can learn how to aggregate by rows or columns. For now, the key point is: a 2D array is still full of numbers, and NumPy can still summarize them.

## 10. Common mistakes and confusion points

### Mistake 1: expecting list behavior
Students often expect array addition to join values together. It does not.

### Mistake 2: forgetting that comparisons return booleans
`temps > 20` does not return the temperatures themselves. It returns `True` and `False`.

### Mistake 3: mixing up an array result and a summary result
- `scores + 1` returns another array
- `scores.sum()` returns one number

These are different kinds of results, and students should learn to read both.
