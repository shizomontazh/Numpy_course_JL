## 2. `np.arange()`: create a sequence with a step

`np.arange(start, stop, step)` works a lot like Python's `range()`, but it creates a NumPy array.

Important idea:
- `start` is included
- `stop` is **not** included


````python
a = np.arange(0, 10, 1)
b = np.arange(2, 15, 3)
c = np.arange(10, 0, -2)

print("a:", a)
print("b:", b)
print("c:", c)
````

## 3. `np.linspace()`: a fixed number of evenly spaced values

`np.linspace(start, stop, num)` is different from `arange()`.

With `linspace`, you decide:
- where the sequence starts
- where it ends
- how many values you want

Important idea:
- by default, the **end value is included**


````python
x = np.linspace(0, 1, 5)
y = np.linspace(10, 20, 6)

print("x:", x)
print("y:", y)
````

This is useful when you want:
- exactly 5 points between 0 and 1
- exactly 8 positions between two limits
- smooth spacing for graphs or experiments

You are controlling the **number of values**, not the step size.


### Compare `arange` and `linspace`

Look at these two lines:

```python
np.arange(0, 1, 0.25)
np.linspace(0, 1, 5)
```

They may look similar, but the logic is different:
- `arange`: choose the step
- `linspace`: choose the number of values


````python
print(np.arange(0, 1, 0.25))
print(np.linspace(0, 1, 5))
````

## 4. Shape matters: from a line to a grid

Very often data starts as one long line of numbers, but we want to view it as a table.

That is what `reshape()` does.

Example:
- 12 values in a row
- reshape into 3 rows and 4 columns


````python
numbers = np.arange(1, 13)
grid = numbers.reshape(3, 4)

print("Flat array:")
print(numbers)
print()
print("Reshaped array:")
print(grid)
````

### Key rule for `reshape`

The total number of elements must stay the same.

So:
- 12 values can become `3 x 4`
- 12 values can become `2 x 6`
- 12 values can become `4 x 3`

But 12 values **cannot** become `5 x 3`, because that would need 15 elements.

## 5. Let NumPy figure out one dimension with `-1`

Sometimes you know one dimension, but not the other.

NumPy can calculate the missing dimension if you use `-1`.


````python
values = np.arange(24)

print(values.reshape(4, 6))
print()
print(values.reshape(6, 4))
print()
print(values.reshape(3, -1))
print()
print(values.reshape(-1, 8))
````

This is very useful, but use it carefully:
- only **one** dimension can be `-1`
- the total number of elements still has to match

## 8. Common mistakes

Here are a few very common problems:

### Mistake 1: wrong reshape size
```python
np.arange(10).reshape(3, 4)
```
This fails because 10 values cannot fill 12 places.

### Mistake 2: forgetting that `stop` is excluded in `arange`
```python
np.arange(1, 5)
```
This gives `[1 2 3 4]`, not `[1 2 3 4 5]`.

### Mistake 3: confusing `arange` and `linspace`
- `arange`: step size
- `linspace`: number of values

### Mistake 4: expecting random arrays to stay the same
Random output changes unless a seed is fixed.
