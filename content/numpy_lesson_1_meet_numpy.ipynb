## 1. Why do we need NumPy?

Python lists are useful, but once we start working with a lot of numbers, tables, grids, pixels, coordinates, or game stats, we want a tool built exactly for that job.

NumPy gives us:

- a special data structure called an **array**
- fast operations on many numbers at once
- a clean way to work with tables and grids

A good first mental model is this:

- a **list** is a general-purpose Python container
- a **NumPy array** is a structure designed for numerical data

## 2. First arrays

The main object in NumPy is the **array**.

We create one with `np.array(...)`.


````python
numbers = np.array([4, 8, 15, 16, 23, 42])
print(numbers)
print(type(numbers))
````

A few things to notice:

- the array was created from a Python list
- the printed result looks list-like, but it is **not** a list
- `type(numbers)` shows that this object belongs to NumPy


### Compare: list vs array


````python
python_list = [4, 8, 15, 16, 23, 42]
numpy_array = np.array([4, 8, 15, 16, 23, 42])

print("python_list:", python_list)
print("numpy_array:", numpy_array)
print()

print("Type of python_list:", type(python_list))
print("Type of numpy_array:", type(numpy_array))
````

Lists and arrays can look similar from the outside, but they are not the same thing.

For now, remember this practical rule:

- use **lists** for general Python work
- use **NumPy arrays** when your data is mostly numbers and you want to analyze or transform it

## 3. Arrays can be one-dimensional or two-dimensional

A 1D array is like a row of values.

A 2D array is like a table or a grid.


````python
one_d = np.array([10, 20, 30, 40])
two_d = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print("1D array:")
print(one_d)
print()
print("2D array:")
print(two_d)
````

This is important because many real problems use 2D structures:

- a scoreboard with rows and columns
- a grid in a game
- a tiny black-and-white image
- a class table with student results

## 4. Looking inside an array

NumPy gives us simple tools for inspecting arrays.

The most useful ones at the start are:

- `shape` — how many rows/columns/etc.
- `ndim` — number of dimensions
- `size` — total number of elements


````python
scores = np.array([
    [12, 15, 18],
    [20, 17, 19]
])

print(scores)
print()
print("shape:", scores.shape)
print("ndim:", scores.ndim)
print("size:", scores.size)
````

Let's decode that:

- `shape` is `(2, 3)` because the array has 2 rows and 3 columns
- `ndim` is `2` because this is a 2D array
- `size` is `6` because there are 6 numbers in total

## 5. Building arrays quickly

Typing every number by hand is fine for tiny examples, but not for everything.

NumPy has helper functions for building arrays quickly.


### `np.arange`

`np.arange(start, stop, step)` works a lot like Python's `range`, but it creates a NumPy array.


````python
a = np.arange(0, 10, 2)
b = np.arange(5)

print("a:", a)
print("b:", b)
````

### `np.zeros` and `np.ones`

These functions create arrays filled with zeros or ones.


````python
zeros_example = np.zeros((2, 3))
ones_example = np.ones((3, 2))

print("zeros_example:")
print(zeros_example)
print()
print("ones_example:")
print(ones_example)
````

These functions are useful when students need to build:

- empty grids
- score tables
- placeholders
- masks and patterns later in the course

## 8. Common mistakes

### Mistake 1: confusing lists and arrays

This is a list:

```python
[1, 2, 3]
```

This is a NumPy array:

```python
np.array([1, 2, 3])
```

### Mistake 2: uneven rows in a 2D array

A 2D array should look like a real table: each row should have the same number of elements.


````python
good = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(good)
````

If rows have different lengths, the result is not the clean 2D structure students usually expect. In an introductory course, it is better to keep rows even and table-like.
