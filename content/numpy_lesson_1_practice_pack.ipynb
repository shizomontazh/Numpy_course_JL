## Part 1. Quick quiz questions

These are intentionally short. They check whether students understand the core language of the lesson before they start writing code.

### [QUIZ] Q1
What is the most common alias used when importing NumPy in Python?

**Expected answer:** `np`

### [QUIZ] Q2
Which line creates a NumPy array?

A. `numbers = [1, 2, 3]`  
B. `numbers = np.array([1, 2, 3])`  
C. `numbers = array(1, 2, 3)`  
D. `numbers = np.list([1, 2, 3])`

**Correct answer:** B

### [QUIZ] Q3
What does `arr.ndim` tell you?

A. The total number of elements  
B. The shape of the array  
C. The number of dimensions  
D. The largest element

**Correct answer:** C

### [QUIZ] Q4
If

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
```

what is `arr.shape`?

A. `(6,)`  
B. `(2, 3)`  
C. `(3, 2)`  
D. `2`

**Correct answer:** B

### [QUIZ] Q5
What does `np.zeros((2, 4))` create?

A. A 1D array with 8 zeros  
B. A 2D array with 2 rows and 4 columns filled with zeros  
C. A 2D array with 4 rows and 2 columns filled with zeros  
D. An empty array

**Correct answer:** B

### [QUIZ] Q6
What is the value of `size` for this array?

```python
arr = np.array([[10, 20], [30, 40], [50, 60]])
```

**Expected answer:** `6`

### [QUIZ] Q7
Which function is the best fit if you want numbers from 0 to 9?

A. `np.ones(10)`  
B. `np.zeros(10)`  
C. `np.arange(10)`  
D. `np.array(10)`

**Correct answer:** C

### [QUIZ] Q8
True or false: a NumPy array can be two-dimensional.

**Correct answer:** True
