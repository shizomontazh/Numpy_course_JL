## Part 1. Quiz-style questions

These questions are good candidates for multiple choice, short answer, or matching tasks on the platform.


### Quiz 1
Given `arr = np.array([5, 10, 15, 20])`, what does `arr[2]` return?

**Answer:** `15`


### Quiz 2
Given `arr = np.array([5, 10, 15, 20])`, what does `arr[-1]` return?

**Answer:** `20`


### Quiz 3
Given `arr = np.array([0, 1, 2, 3, 4, 5])`, what does `arr[1:4]` return?

**Answer:** `array([1, 2, 3])`


### Quiz 4
Given `arr = np.array([0, 1, 2, 3, 4, 5])`, what does `arr[:3]` return?

**Answer:** `array([0, 1, 2])`


### Quiz 5
Given `arr = np.array([0, 1, 2, 3, 4, 5])`, what does `arr[::2]` return?

**Answer:** `array([0, 2, 4])`


### Quiz 6
Given
```python
matrix = np.array([
    [10, 11, 12],
    [20, 21, 22],
    [30, 31, 32]
])
```
what does `matrix[1, 2]` return?

**Answer:** `22`


### Quiz 7
Using the same matrix, what does `matrix[0]` return?

**Answer:** `array([10, 11, 12])`


### Quiz 8
Using the same matrix, what does `matrix[:, 1]` return?

**Answer:** `array([11, 21, 31])`


### Quiz 9
Using the same matrix, what does `matrix[1:, :2]` return?

**Answer:**
```python
array([
    [20, 21],
    [30, 31]
])
```


### Quiz 10
What is the main difference between `matrix[1, 2]` and `matrix[1:2, 2:3]`?

**Answer:**  
The first returns one value.  
The second returns a subarray.
